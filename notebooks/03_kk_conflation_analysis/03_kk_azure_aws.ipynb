```python
base_path = '../../data'
```

```python
azure_aws = pd.read_csv(f'{base_path}/processed/azure_aws_dedupe/data_matching_output.csv')
```

```python
azure_aws = azure_aws.sort_values(['Cluster ID', 'source file'])
azure_aws[azure_aws['Link Score'] > .30]
```

```python
# azure_aws[azure_aws['name'] == 'Azure Cosmos DB']
azure_aws[azure_aws['name'].isin(['Azure Functions', 'AWS Lambda'])].to_dict(orient='records')
```

```python
def add_azure_to_name(row):
    if row['source file'] == 0:
        if 'Azure' not in row['name']:
            row['name'] = f"Azure {row['name']}"
    return row
    
azure_aws = azure_aws.apply(add_azure_to_name, axis=1)
```

```python
for prop in ['name', 'categories', 'short_description', 'long_description']:
    azure_aws.loc[(azure_aws['Cluster ID'] == 6) & (azure_aws['source file'] == 1), prop] = list(azure_aws[azure_aws['Cluster ID'] == 175][prop])[0]
azure_aws.loc[(azure_aws['Cluster ID'] == 6) & (azure_aws['source file'] == 1), prop] 
```
